### 1. Задание
Даны два вектора в трехмерном пространстве: $(10,10,10)$ и $(0,0,-10)$

1)	Найдите их сумму. (на листочке)  
2)	Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
import numpy as np
x1 = np.array([10, 10, 10])
x2 = np.array([0, 0, -10])
```

```python
def len_vector(x):
    len_vector = (x[0]**2 + x[1]**2 + x[2]**2)**(1/2)
    return len_vector
```

```python
print('Длина вектора x1:', len_vector(x1))
print('Длина вектора x2:', len_vector(x2))
```

```python
# гипербола
a = 10
b = 3
x = np.arange(-a, a+0.1, 0.1)
y1 = np.sqrt((b**2 * (a**2 + x**2))/a**2)
y2 = -np.sqrt((b**2 * (a**2 + x**2))/a**2)

plt.plot(x, y1)
plt.plot(x, y2)
```
