### 4. Задание (в программе):
Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams["figure.figsize"] = 10, 5
```

```python
x = np.linspace(0, 10, 160)
k1 = 1
k2 = 2.4
plt.plot(x, np.cos(k1*x), label = "k1 = 1")
plt.plot(x, np.cos(k2*x), label = "k2 = 2.4" )
plt.title("График функции y(k,x)=cos(k∙x)", fontsize=16)
plt.legend(frameon=False);
```
